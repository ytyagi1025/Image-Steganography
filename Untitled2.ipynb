```
from keras.datasets import mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```
import keras
from keras import Model
from keras.layers import Conv2D,Flatten,Dense,Input
```

```
x1=Input(shape=(28,28,1))
x=Conv2D(16,(3,3),activation="relu",strides=2)(x1)
x=Conv2D(8,(3,3),activation="relu",strides=2)(x)
x=Conv2D(3,(3,3),activation="relu",strides=2)(x)
x=Flatten()(x)
o=Dense(10,activation="sigmoid")(x)
cls=Model(x1,o)
cls.summary()
```

```
cls.compile(optimizer="Adam",loss="mse")
```

```
x_train=x_train.reshape(60000,28,28,1)
x_train=x_train/255
#y_train=y_train.reshape(60000,28,28,1)
from keras.utils import to_categorical
num_classes = 10
y_train = keras.utils.to_categorical(y_train, num_classes)

print(x_train.shape)
```

```
print(y_train.shape)
```

```
cls.fit(x_train,y_train,epochs=10)
```

```
#x_test=x_test/255
x_test=x_test.reshape(10000,28,28,1)
#y_test = keras.utils.to_categorical(y_test, num_classes)
score = cls.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score)
#print('Test accuracy:', score[1])
```
