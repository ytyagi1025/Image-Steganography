```
import tensorflow as tf
import numpy, cv2 as cv
```

```
from google.colab import drive
drive.mount('/content/drive')
```

```
src_test1=[]
src_test2=[]
for i in range(100):
  src_test1.append(cv.imread("/content/drive/My Drive/xtestset/{0}.jpeg".format("img"+str(i))))
  src_test2.append(cv.imread("/content/drive/My Drive/hide/{0}.jpeg".format("img"+str(i)+"in")))
xtestset=numpy.asarray(src_test1)
decoded_imgs=numpy.asarray(src_test2)
```

```
print(xtestset.shape)
```

```
file1 = open("/content/drive/My Drive/psnrssim.txt","w")
avgs=0
avgp=0
avgm=0
#sess=tf.InteractiveSession()
sess=tf.compat.v1.InteractiveSession()
for i in range(0,100):
    ssim1 = tf.image.ssim(tf.convert_to_tensor(xtestset[i]),tf.convert_to_tensor(decoded_imgs[i]),max_val=1).numpy()
    file1.write(str(i)+". = \t")
    file1.write("SSIM:\t"+str(ssim1))
    print(ssim1)
    t=tf.image.psnr(tf.convert_to_tensor(xtestset[i]),tf.convert_to_tensor(decoded_imgs[i]),max_val=1).numpy()
    file1.write("\tPSNR:\t"+str(t))
    t1=tf.image.ssim_multiscale(tf.convert_to_tensor(decoded_imgs[i]),tf.convert_to_tensor(xtestset[i]),max_val=1).numpy()
    file1.write("\tMs-ssim:\t"+str(t1)+"\n")
    avgs+=ssim1
    avgm+=t1
    print(t)
    print(t1)
    avgp+=t
    print(i)
file1.write("AVERAGE. = \t")
file1.write("SSIM:\t"+str(avgs/100))
file1.write("MS-SSIM:\t"+str(avgm/100))
file1.write("PSNR:\t"+str(avgp/100)+"\n")
print("avg=",avgs/100,avgp/100,avgm/100)
file1.close()
sess.close()
```
